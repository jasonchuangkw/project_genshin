<b>Info and Stats Section</b>

|Field|Description|Data Type|Restriction|
|:--|:--|:--|:--|
|uid||string|-|
|hoyolab_uid||string|-|
|server|server which the player is based|string (categorical)|asia,|
|level|Level of player (range between 0 to 60)|numeric|range from 1 to 60|

```python
#prepare df column names from csv (varaible_characters) for Character
dict_infostats = pd.read_csv('data_setup//variable_infostats.csv', header=None, index_col=0).to_dict() #read csv to dictionary
df_infostats_default = pd.DataFrame.from_dict(dict_infostats, orient ='index')
```

```python
#function to convert data from info and stats section to df

def data_infostats_input(i, data):
    
    
    df_infostats = df_infostats_default.copy(deep=True)
    df_infostats.uid = i
    df_infostats.level = data.info.level
    df_infostats.nickname = data.info.nickname
    df_infostats.server = data.info.server
    df_infostats.achievements = data.stats.achievements
    df_infostats.days_active = data.stats.days_active
    df_infostats.characters = data.stats.characters
    df_infostats.spiral_abyss = data.stats.spiral_abyss   
    df_infostats.anemoculi = data.stats.anemoculi  
    df_infostats.geoculi = data.stats.geoculi     
    df_infostats.dendroculi = data.stats.dendroculi    
    df_infostats.electroculi = data.stats.electroculi
    df_infostats.common_chests = data.stats.common_chests
    df_infostats.exquisite_chests = data.stats.exquisite_chests
    df_infostats.precious_chests = data.stats.precious_chests
    df_infostats.luxurious_chests = data.stats.luxurious_chests
    df_infostats.remarkable_chests = data.stats.remarkable_chests
    df_infostats.unlocked_waypoints = data.stats.unlocked_waypoints
    df_infostats.unlocked_domains = data.stats.unlocked_domains


    return df_infostats
```

<b>Explorations Section</b>

```python
#prepare df column names from csv (varaible_characters) for Explorations
dict_explorations = pd.read_csv('data_setup//variable_explorations.csv', header=None, index_col=0).to_dict() #read csv to dictionary
df_explorations_default = pd.DataFrame.from_dict(dict_explorations, orient ='index')
```

```python
#function to convert data from Explorations Section

def data_explorations_input(i, data):
    
    df_explorations = df_explorations_default.copy(deep=True)
    df_explorations.uid = i
    
    n = 0
    
    for i in data.explorations:
        match i.id:
            case 1: #Mondstadt
                df_explorations.mondstadt_explored = i.raw_explored
                df_explorations.mondstadt_reputation = i.level 
                
            case 2: #Liyue
                df_explorations.liyue_explored = i.raw_explored                
                df_explorations.liyue_reputation = i.level 
                
            case 3: #Dragonspine
                df_explorations.dragonspine_explored = i.raw_explored
                df_explorations.dragonspine_offering = i.level 
                df_explorations.dragonspine_tree_level = i.offerings[0].level
                
            case 4: #Inazuma
                df_explorations.inazuma_explored = i.raw_explored
                df_explorations.inazuma_reputation = i.level
                df_explorations.inazuma_tree_level = i.offerings[0].level
                
            case 5: #Enkanomiya
                df_explorations.enkanomiya_explored = i.raw_explored
                df_explorations.enkanomiya_offering = i.level

            case 6: #The Chasm
                df_explorations.the_chasm_explored = i.raw_explored
                df_explorations.the_chasm_offering = i.level
                the_chasm_lumenstone_level = i.offerings[0].level

            case 7: #The Chasm: Underground Mines
                df_explorations.underground_mines_explored = i.raw_explored
                df_explorations.underground_mines_offering = i.level
                df_explorations.underground_lumenstone_level = i.offerings[0].level

            case 8: #Sumeru
                df_explorations.sumeru_explored = i.raw_explored
                df_explorations.sumeru_reputation = i.level
                df_explorations.sumeru_tree_level = i.offerings[0].level
                
        n += 1    
        
    return df_explorations
```

```python
def data_teapot_input(i, data):
    
    df_teapot = df_teapot_default.copy(deep=True)
    df_teapot.uid = i
    df_teapot.teapot_level = data.teapot.level
    df_teapot.teapot_visitors = data.teapot.visitors
    df_teapot.teapot_comfort = data.teapot.comfort
    df_teapot.teapot_items = data.teapot.items
    
    for i in data.teapot.realms:
            
        match i.name:
            case 'Floating Abode': #Mondstadt
                df_teapot.floating_abode = 1
                
            case 'Emerald Peak': #Mondstadt
                df_teapot.emerald_peak = 1  
                
            case 'Cool Isle': #Mondstadt
                df_teapot.cool_isle = 1  

            case 'Silken Courtyard': #Mondstadt
                df_teapot.silken_courtyard = 1  

            case 'Sublime Spicewood': #Mondstadt
                    df_teapot.sublime_spicewood = 1  

    return df_teapot
```

```python
def data_abyss_input(i, data):
    
    df_abyss = df_abyss_default.copy(deep=True)
    df_abyss.uid = i
    
    if data.abyss.current.unlocked == True:
        df_abyss.abyss_c_unlocked = 1
        
    df_abyss.abyss_c_max_floor = data.abyss.current.max_floor
    df_abyss.abyss_c_total_stars = data.abyss.current.total_stars

    
    if data.abyss.previous.unlocked == True:
        df_abyss.abyss_p_unlocked = 1

    df_abyss.abyss_c_max_floor = data.abyss.previous.max_floor
    df_abyss.abyss_c_total_stars = data.abyss.previous.total_stars

    #most played character 1,2,3 and 4 in current abyss
    n = 0
    for i in data.abyss.current.ranks.most_played:
        
        if i == 0:
            df_abyss.abyss_c_mp_name_1 =  i.name
            df_abyss.abyss_c_mp_rarity_1 =  i.rarity
            df_abyss.abyss_c_mp_value_1 =  i.value
            
        if i == 1:
            df_abyss.abyss_c_mp_name_2 =  i.name
            df_abyss.abyss_c_mp_rarity_2 =  i.rarity
            df_abyss.abyss_c_mp_value_2 =  i.value
            
        if i == 2:
            df_abyss.abyss_c_mp_name_3 =  i.name
            df_abyss.abyss_c_mp_rarity_3 =  i.rarity
            df_abyss.abyss_c_mp_value_3 =  i.value
            
        if i == 3:
            df_abyss.abyss_c_mp_name_4 =  i.name
            df_abyss.abyss_c_mp_rarity_4 =  i.rarity
            df_abyss.abyss_c_mp_value_4 =  i.value
            
        n += 1  
        
    #most played character 1,2,3 and 4 in previous abyss    
    n = 0
    for i in data.abyss.previous.ranks.most_played:
        if n == 0:
            df_abyss.abyss_p_mp_name_1 =  i.name
            df_abyss.abyss_p_mp_rarity_1 =  i.rarity
            df_abyss.abyss_p_mp_value_1 =  i.value
            
        if i == 1:
            df_abyss.abyss_p_mp_name_2 =  i.name
            df_abyss.abyss_p_mp_rarity_2 =  i.rarity
            df_abyss.abyss_p_mp_value_2 =  i.value
            
        if i == 2:
            df_abyss.abyss_p_mp_name_3 =  i.name
            df_abyss.abyss_p_mp_rarity_3 =  i.rarity
            df_abyss.abyss_p_mp_value_3 =  i.value
            
        if i == 3:
            df_abyss.abyss_p_mp_name_4 =  i.name
            df_abyss.abyss_p_mp_rarity_4 =  i.rarity
            df_abyss.abyss_p_mp_value_4 =  i.value
            
        n += 1
        
    #most kill character in current abyss
    try:
        df_abyss.abyss_c_mk_name =  data.abyss.current.ranks.most_kills[0].name
        df_abyss.abyss_c_mk_rarity =  data.abyss.current.ranks.most_kills[0].rarity
        df_abyss.abyss_c_mk_value =  data.abyss.current.ranks.most_kills[0].value
    except:
        pass

    #most kill character in previous abyss     
    try:    
        df_abyss.abyss_p_mk_name =  data.abyss.previous.ranks.most_kills[0].name
        df_abyss.abyss_p_mk_rarity =  data.abyss.previous.ranks.most_kills[0].rarity
        df_abyss.abyss_p_mk_value =  data.abyss.previous.ranks.most_kills[0].value   
    except:
        pass

    #most strongest strike character in current abyss
    try:
        df_abyss.abyss_c_ss_name =  data.abyss.current.ranks.strongest_strike[0].name
        df_abyss.abyss_c_ss_rarity =  data.abyss.current.ranks.strongest_strike[0].rarity
        df_abyss.abyss_c_ss_value =  data.abyss.current.ranks.strongest_strike[0].value
    except:
        pass

    #most strongest strike character in previous abyss     
    try:    
        df_abyss.abyss_p_ss_name =  data.abyss.previous.ranks.strongest_strike[0].name
        df_abyss.abyss_p_ss_rarity =  data.abyss.previous.ranks.strongest_strike[0].rarity
        df_abyss.abyss_p_ss_value =  data.abyss.previous.ranks.strongest_strike[0].value   
    except:
        pass

    #most most damage taken character in current abyss
    try:
        df_abyss.abyss_c_mdt_name =  data.abyss.current.ranks.most_damage_taken[0].name
        df_abyss.abyss_c_mdt_rarity =  data.abyss.current.ranks.most_damage_taken[0].rarity
        df_abyss.abyss_c_mdt_value =  data.abyss.current.ranks.most_damage_taken[0].value
    except:
        pass

    #most most damage taken character in previous abyss     
    try:    
        df_abyss.abyss_p_mdt_name =  data.abyss.previous.ranks.most_damage_taken[0].name
        df_abyss.abyss_p_mdt_rarity =  data.abyss.previous.ranks.most_damage_taken[0].rarity
        df_abyss.abyss_p_mdt_value =  data.abyss.previous.ranks.most_damage_taken[0].value   
    except:
        pass
    
    #most most burst used character in current abyss
    try:
        df_abyss.abyss_c_mbu_name =  data.abyss.current.ranks.most_bursts_used[0].name
        df_abyss.abyss_c_mbu_rarity =  data.abyss.current.ranks.most_bursts_used[0].rarity
        df_abyss.abyss_c_mbu_value =  data.abyss.current.ranks.most_bursts_used[0].value
    except:
        pass

    #most most burst used character in previous abyss     
    try:    
        df_abyss.abyss_p_mbu_name =  data.abyss.previous.ranks.most_bursts_used[0].name
        df_abyss.abyss_p_mbu_rarity =  data.abyss.previous.ranks.most_bursts_used[0].rarity
        df_abyss.abyss_p_mbu_value =  data.abyss.previous.ranks.most_bursts_used[0].value   
    except:
        pass    
    
    #number of stars on each floor (current abyss)  
    try: 
        df_abyss.abyss_c_stars_floor9 = data.abyss.current.floors[0].stars
        df_abyss.abyss_c_stars_floor10 = data.abyss.current.floors[1].stars
        df_abyss.abyss_c_stars_floor11 = data.abyss.current.floors[2].stars
        df_abyss.abyss_c_stars_floor12 = data.abyss.current.floors[3].stars
        
    except:
        pass 

    #number of stars on each floor (previous abyss) 
    try: 
        df_abyss.abyss_p_stars_floor9 = data.abyss.previous.floors[0].stars
        df_abyss.abyss_p_stars_floor10 = data.abyss.previous.floors[1].stars
        df_abyss.abyss_p_stars_floor11 = data.abyss.previous.floors[2].stars
        df_abyss.abyss_p_stars_floor12 = data.abyss.previous.floors[3].stars
        
    except:
        pass

    return df_abyss
```
